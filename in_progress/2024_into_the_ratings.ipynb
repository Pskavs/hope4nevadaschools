What are these Columns?
---
There are 34 Columns that are in our data. In order to figure out what we need, I will list the columns in the following categories: 1) Critical: Need these columns for almost all analysis. These columns include mostly quantitative data. 2) Important: Might include categorical information that includes specific information about a school (Such as is it private / charter) 3) Niche Specific: These columns are only important to certain categories of school (such as high school graduation, or credit sufficiency) 4) Low importance: Won't be used for the data analysis, but might be useful to know. This includes school designations and participation categories.

**The Critical columns are as follows:**
1. School Name
2. Total Index Score (This is the score given to the school based on other factors).
3. Star Rating (Based on their TIS), Title 1 Status (Title I are schools in low socio-economic neighborhoods that are given extra funding)
4. % Proficient ELA (These three proficiencies are the percent of students considered "Proficient" based on SBAC testing)
5. % Proficient Math
6. % Proficient Science
7. % Meeting AGP ELA (AGP stands for adequate growth percentile, or how many students are achieving adequate growth)
8. % Meeting AGP Math
9. Perc_WIDA_AGP (WIDA testing is a subcategory of tests given to English Language Learners)
10. % Meeting AGP EL (Separate category for EL students to see if they are meeting growth).
11. Chronic Absenteeism (% of students missing more than 10% of days)

**The important columns are as follows:**
1. District Code (Allows us to easily classify which district we are working with)
2. School Type (Tells us whether the school is a public, magnet, or charter school)

**The niche specific columns are:**
1. NAC 389.445 Credit Requirements (8th Grade): A law that states an 8th-grade student must earn 1.5 units of credit in English, 1.5 units of credit in mathematics, 1 unit of credit in science, and 1 unit of credit in social studies.
2. 9th Grade Credit Sufficiency (Makes sure that 9th graders have earned at least 5 credits at the end of their first year in high school)
3. CCR Participation: Percent of students who participate in a College and Career Readiness Program including - Advanced Placement (AP), International Baccalaureate (IB), Dual Credit/Dual Enrollment (DC/DE), and Career and Technical Education (CTE)
4. CCR Completion: Percent of students who finish the CCR program.
5. Advanced Diploma: Percent of students who receive the Advanced Diploma. Requires 3.25 GPA, 4 years of math, 3 years of science, and 24 credits.
6. 4 Year Graduation Rate
7. 5 Year Graduation Rate

**The low importance category include:**
1. School Code
2. 95% Participation Met Mathematics: ALl of the participation categories involve tracking participation in state testing. I will most likely use these columns to remove schools that might be outliers.
3. 95% Participation Met ELA
4. Participation Warning
5. Participation Penalty
6. Continuing Participation Penalty
7. CSI: This designation is for the lowest 5 % of schools in CCSD
8. Yrs In Designation As CSI
9. TSI: Designation is for schools that have minority subgroups that are consistently under performing.
10. Yrs in Designation As TSI
11. ATSI: Designation for schools that fit the TSI category, but it is bad enough that the subgroup on its own is the cause of the school being CSI.
12. Yrs in Designation as ATSI.
